Matplotlib is a Python 2D plotting library which produces publication quality figures in a variety of hardcopy formats and interactive environments across platforms. Matplotlib can be used in Python scripts, the Python and IPython shells, the Jupyter notebook, web application servers, and four graphical user interface toolkits.
Refer to https://matplotlib.org/ for more detail and click 'library' section.

There are two ways of creating matplotlib plots
 - Functional method
 - Object Oriented method

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
%matplotlib inline
```

```python
# Above line allows to see the plot inside the jupyter notebook. 
#But, if you are not on jupyternote book then do plt.show() in the end of code to display the plot. 
#plt.show() prints the plot instead of just showing the plot.
```

```python
Data = pd.read_excel('Sample.xlsx',sheetname='Sheet1')
```

```python
Data.head()
```

## The functional method

```python
plt.plot(Data['ABC'], Data['XYZ'])
```

#### Adding x, y label and title 

```python
plt.plot(Data['ABC'], Data['XYZ'])
plt.xlabel('ABC')
plt.ylabel('XYZ')
plt.title('Relation between ABC and XYZ')
```

#### Creating multiiple plots together

For this I use subplot() function

```python
plt.subplot(1, 2, 1) # 1 -> Number of rows, 2 -> Number of Columns, 1 -> Plot number
plt.plot(Data['ABC'], Data['XYZ'], 'b')

plt.subplot(1, 2, 1)
plt.plot(Data['XYZ'], Data['ABC'], 'r')
```

```python
plt.subplot(1, 2, 1) # 1 -> Number of rows, 2 -> Number of Columns, 1 -> Plot number
plt.plot(Data['ABC'], Data['XYZ'], 'b')
plt.xlabel('ABC')
plt.ylabel('XYZ')
plt.title('Relation between ABC and XYZ')

plt.subplot(1, 2, 2)
plt.plot(Data['XYZ'], Data['ABC'], 'r')
plt.ylabel('ABC')
plt.xlabel('XYZ')
plt.title('Relation between XYZ and ABC')
```

## The Object Oriented method 

```python
fig = plt.figure() # creating blank figure


axes1 = fig.add_axes([0.1, 0.1, 0.9, 0.9]) 
axes2 = fig.add_axes([0.2, 0.3, 0.4, 0.5]) # Left, bottom, width, and height.
```

```python
fig = plt.figure()


axes1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])
axes2 = fig.add_axes([0.2, 0.3, 0.4, 0.5])

axes1.plot(Data['ABC'], Data['XYZ'], 'b')
axes2.plot(Data['XYZ'], Data['ABC'], 'r')
```

Now, I want to move my inner plot to lower right corner to make clear.

```python
fig = plt.figure()


axes1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])
axes2 = fig.add_axes([0.2, 0.11, 0.4, 0.5]) # I changed 0.3 to 0.11

axes1.plot(Data['ABC'], Data['XYZ'], 'b')
axes2.plot(Data['XYZ'], Data['ABC'], 'r')
```

```python
fig = plt.figure()


axes1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])
axes2 = fig.add_axes([0.6, 0.11, 0.4, 0.5]) # I changed 0.2 to 0.6

axes1.plot(Data['ABC'], Data['XYZ'], 'b')
axes2.plot(Data['XYZ'], Data['ABC'], 'r')
```

Lets label the figure

```python
fig = plt.figure()


axes1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])
axes2 = fig.add_axes([0.6, 0.11, 0.4, 0.5]) # I changed 0.2 to 0.6

axes1.plot(Data['ABC'], Data['XYZ'], 'b')
axes1.set_title('Outer Plot')

axes2.plot(Data['XYZ'], Data['ABC'], 'r')
axes2.set_title('Inner Plot')
```

#### Working with Subplot 

```python
fig, axes = plt.subplots(1, 2)

'''
The axes object is just the list of matplotlib axes object. 
we can loop through those object. 
'''
```

```python
fig, axes = plt.subplots(1, 2)

axes[0].plot(Data['ABC'], Data['XYZ'])
axes[1].plot(Data['XYZ'], Data['ABC'])

#plt.tight_layout() # This removes overlapping in plots
```

## Figure Size and DPI(dot per inch; Pixel per inch)

```python
fig = plt.figure(figsize = (1, 2)) # 1 inch, 2 inch

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(Data['ABC'], Data['XYZ'])
```

```python
fig = plt.figure(figsize = (6, 2))

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(Data['ABC'], Data['XYZ'])
```

### Adding legend in plot

```python
plt.subplot(1, 2, 1)
plt.plot(Data['ABC'], Data['XYZ'], label = 'ABCXYZ')

plt.subplot(1, 2, 1)
plt.plot(Data['XYZ'], Data['ABC'], label = 'XYZABC')
# loc deterimines the position of legend on the plot. It varies
# from 0 to 10. Google it for more detail.
plt.legend(loc = 0)
```

### Color and Line thickness 

```python
fig, axes = plt.subplots(1, 2)
# To set up transperacny use alpha parameter
axes[0].plot(Data['ABC'], Data['XYZ'], color = 'green', linewidth = 10)
axes[1].plot(Data['XYZ'], Data['ABC'], color = 'red', linewidth = 5)
```

```python
fig, axes = plt.subplots(1, 2)
# To set up transperacny use alpha parameter
axes[0].plot(Data['ABC'], Data['XYZ'], color = 'green', alpha = 0.6, linewidth = 10)
axes[1].plot(Data['XYZ'], Data['ABC'], color = 'red', alpha = 0.4, linewidth = 5)
```

```python
fig, axes = plt.subplots(1, 2)
# To set up transperacny use alpha parameter
axes[0].plot(Data['ABC'], Data['XYZ'], color = 'green', alpha = 0.6, linestyle = '--',linewidth = 10)
axes[1].plot(Data['XYZ'], Data['ABC'], color = 'red', alpha = 0.4, linestyle = '-.',linewidth = 5)
```

```python
fig, axes = plt.subplots(1, 2)
# To set up transperacny use alpha parameter
axes[0].plot(Data['ABC'], Data['XYZ'], color = 'green', alpha = 0.6, linestyle = ':',linewidth = 10)
axes[1].plot(Data['XYZ'], Data['ABC'], color = 'red', alpha = 0.4, linestyle = 'steps',linewidth = 5)
```

```python
fig, axes = plt.subplots(1, 2)
# To set up transperacny use alpha parameter
axes[0].plot(Data['ABC'], Data['XYZ'], color = 'green', alpha = 0.6, ls = ':',lw = 10)
axes[1].plot(Data['XYZ'], Data['ABC'], color = 'red', alpha = 0.4, ls = 'steps',lw = 5)
```

```python
fig, axes = plt.subplots(1, 2)
# To set up transperacny use alpha parameter
axes[0].plot(Data['ABC'], Data['XYZ'], color = 'green', lw = 1, marker = 'o' )
axes[1].plot(Data['XYZ'], Data['ABC'], color = 'red', lw = 1, marker = '1')
```

```python
fig, axes = plt.subplots(1, 2)
# To set up transperacny use alpha parameter
axes[0].plot(Data['ABC'], Data['XYZ'], color = 'green', lw = 10, marker = 'o' )
axes[1].plot(Data['XYZ'], Data['ABC'], color = 'red', lw = 5, marker = '1')

"""
Marker disappeared because of high linewidth. You can customize color, 
size, style of marker. Just bit it, googl it.
"""
```
